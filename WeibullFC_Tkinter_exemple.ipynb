```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
import predictr
from predictr import Analysis
import matplotlib
matplotlib.use('TkAgg')
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk)
```

```python
BETA = []
ETA = []
metodo = []


window = tk.Tk()
window.title('WeibullFC')

def sel():
    selection = int(str(vR.get()))
    if selection == 1:
        metodo.clear()
        metodo.append('mrr') 
    elif selection == 2:
        metodo.clear()
        metodo.append('mle')

def simple_weibull():
    try:
        text_from_Box = Text_Entry.get("1.0", 'end-1c').split("\n")
        failures = [float(x) for x in text_from_Box]       

        text_from_Box2 = Text_Entry2.get("1.0", 'end-1c').split("\n")
        suspensions = [float(x) for x in text_from_Box2]     

        x = Analysis(df=failures, ds=suspensions, show=True)

        figure = Figure(figsize=(6, 6), dpi=100) 
        axes = figure.add_subplot(111)    

        if metodo[0] == 'mrr':
            x.mrr()
            BETA.clear()
            BETA.append(f'{x.beta:,.3f}')
            ETA.clear()
            ETA.append(f'{x.eta:,.3f}')
            
            Label(window, text = f"Beta: {BETA[0]}  Eta: {ETA[0]}").grid(row = 2, column = 0, padx= 100, sticky = W)

            
        elif metodo[0] == 'mle':
            x.mle()
            BETA.clear()
            BETA.append(f'{x.beta:,.3f}')
            ETA.clear()
            ETA.append(f'{x.eta:,.3f}')
            
            Label(window, text = f"Beta: {BETA[0]}  Eta: {ETA[0]}").grid(row = 2, column = 0, padx= 100, sticky = W)

        elif metodo == '':
            print('metodo est√° nulo')

        #BETA = f'{x.beta:,.3f}'

        #Label(window, text = f'Beta: {x.beta:,.3f} Eta: {x.eta:,.3f}').grid(row = 0, column = 2, sticky = W)
    except:
        print('Nenhum dado inserido...') 
    



Label(window, text = "Failures").grid(row = 0, padx = 10, sticky = W)
Text_Entry = Text(window, height = 30, width = 30)
Text_Entry.grid(row = 1, column = 0, padx = 10)

Label(window, text = "Suspensions").grid(row = 0, column = 1, sticky = W)
Text_Entry2 = Text(window, height = 30, width = 30)
Text_Entry2.grid(row = 1, column = 1)

vR = IntVar()
R1 = Radiobutton(window, text="MRR (Medan Rank Regression)", variable=vR, value=1, command=sel)
R1.grid( row = 0, column = 2, sticky = W)

R2 = Radiobutton(window, text="MLE (Maximum Likelihood Estimator)", variable=vR, value=2, command=sel)
R2.grid( row = 1, column = 2, sticky = W )

Button(window, text = 'Analysis', command = simple_weibull).grid(row = 2, column = 0, padx = 10, pady = 10, sticky = W)




     
    


    
    



















mainloop()
```
