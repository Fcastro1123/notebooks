```python
from kafka import KafkaConsumer
```

```python
broker = 'ec2-3-20-182-100.us-east-2.compute.amazonaws.com:9092'
```

```python
topic_name = 'Rede_5ab63ad6-5e6d-4742-b04f-695a981e9ec4'
```

```python
while True:
    consumer = KafkaConsumer(topic_name,
                             group_id = None,
                             bootstrap_servers=[broker])
    
    for message in consumer:
        print (message.value)
```
